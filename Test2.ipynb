```python
def brute(alist,esum):
    flist=[]
    indlist=[]
    newset=True
    for i in range(len(alist)):
        for j in range(len(alist)):
            for k in range(len(alist)):
                if(i!=j) and (j!=k) and (i!=k):
                    if(alist[i]+alist[j]+alist[k])==esum:
                        if indlist==[]:
                            flist.append([alist[i],alist[j],alist[k]])
                            indlist.append([i,j,k])
                        else:
                            for index in range(len(indlist)):
                                if sorted([i,j,k])!=sorted(indlist[index]):
                                    newset=True
                                else:
                                    newset=False
                                    break
                            if newset==True:
                                flist.append([alist[i],alist[j],alist[k]])
                                indlist.append([i,j,k])
    return(flist,indlist)
```

```python
def brutein(alist,esum):
    flist=[]
    indlist=[]
    newset=True
    for i in range(len(alist)):
        for j in range(len(alist)):
            if i!=j and (esum-alist[i]-alist[j]) in alist:
                return (alist[i],alist[j],(esum-alist[i]-alist[j]))
    return ('Not found')
```

```python
def selector(alist,esum):
    ilist=sorted(alist)
    if ilist[0]>esum/3:
        return('Not found, sum too small')
    elif ilist[-1]<esum/3:
        return('Not found, sum too big')
    i=0
    elif ilist[-1]<esum/2:
        while ilist[i]<(esum-ilist[-1]-ilist[-2]):
            ilist.pop(0)
    elif ilist[-1]>esum:
        i=len(ilist)-1
        while ilist[i]>(esum-ilist[0]-ilist[1]):
            ilist.pop()
            i-=1
    return(len(ilist))
```
