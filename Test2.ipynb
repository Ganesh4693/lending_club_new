```python
def brute(alist,esum):
    flist=[]
    indlist=[]
    newset=True
    for i in range(len(alist)):
        for j in range(len(alist)):
            for k in range(len(alist)):
                if(i!=j) and (j!=k) and (i!=k):
                    if(alist[i]+alist[j]+alist[k])==esum:
                        if indlist==[]:
                            flist.append([alist[i],alist[j],alist[k]])
                            indlist.append([i,j,k])
                        else:
                            for index in range(len(indlist)):
                                if sorted([i,j,k])!=sorted(indlist[index]):
                                    newset=True
                                else:
                                    newset=False
                                    break
                            if newset==True:
                                flist.append([alist[i],alist[j],alist[k]])
                                indlist.append([i,j,k])
                k+=1
            j+=1
        i+=1
    return(flist,indlist)
```
